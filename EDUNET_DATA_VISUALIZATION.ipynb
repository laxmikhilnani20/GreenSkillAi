```python
import matplotlib.pyplot as plt

# Define energy sources and their respective consumption values
energy_sources = ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"]
energy_values = [1200, 1300, 1100, 1500, 1400, 1600]  # Example energy consumption values in MWh

# Pie chart to show energy consumption distribution
plt.pie(energy_values, labels=energy_sources, autopct='%1.1f%%',
        colors=['gold', 'lightblue', "lightgreen", 'coral', 'lightpink', 'lavender'])

# Add a title
plt.title("Energy Consumption Share by Source")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
energy_sources = ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"]
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]  # in MWh
costs = [200, 400, 600, 150, 250, 700]  # in Million $

# Scatter plot: Energy Consumption vs Cost
plt.scatter(energy_consumption, costs, color='blue', alpha=0.7)

# Add labels and title
plt.xlabel("Energy Consumption (MWh)")
plt.ylabel("Cost (Million $)")
plt.title("Energy Consumption vs Cost by Source")

# Annotate each point with energy source
for i, source in enumerate(energy_sources):
    plt.annotate(source, (energy_consumption[i], costs[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Energy consumption data
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]  # in MWh

# Create histogram
plt.hist(energy_consumption, bins=5, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel("Energy Consumption (MWh)")
plt.ylabel("Frequency")
plt.title("Distribution of Energy Consumption")

# Show the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
energy_sources = ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"]
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]  # Energy consumption in MWh

# Create line chart
plt.plot(energy_sources, energy_consumption, marker='o', linestyle='-', color='blue', label='Energy Consumption')

# Add labels and title
plt.xlabel("Energy Sources")
plt.ylabel("Energy Consumption (MWh)")
plt.title("Energy Consumption by Source")

# Add grid and legend
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "Solar": [1200, 1300, 1100, 1500, 1400, 1600],
    "Wind": [3400, 3600, 3200, 3800, 3700, 3900],
    "Hydropower": [2900,3100,2000,2500,2700,3000],
    "Biomass" : [2500, 2700, 2400, 2800, 2600, 2900]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
sns.pairplot(df)
plt.show()
```

```python
df.corr()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show('Correaltion Between Energy Sources')
plt.show()
```
